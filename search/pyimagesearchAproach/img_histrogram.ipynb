```python
import cv2
import glob
import numpy as np
import glob
from matplotlib import pyplot as plt
```

```python

images = list(cv2.imread(file) for file in glob.glob("./images/*.jpg"))

for img_no, img in enumerate(images):
    img= cv2.resize(img,(300,300))
    img= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.show()
```

```python
colors=('r','g','b')
for img in images:
    img= cv2.resize(img,(300,300))
    img= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    for chnl_no,chnl_clr in enumerate(colors):
        histrogam = cv2.calcHist([img], [chnl_no],None,[256],[0,256])
        plt.subplot(1,2,1)
        plt.plot(histrogam,color = chnl_clr)
        plt.xlim([0,256])
    plt.subplot(1,2,2)
    plt.imshow(img)
    plt.show()
```
